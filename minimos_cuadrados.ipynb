```python
import sympy

x_coords = [1, 2, 3, 4, 5]
y_coords = [2.1, 3.9, 6.2, 8.0, 10.1]

# 2. Definir variables simbólicas
x, m, b = sympy.symbols('x m b')

# 3. Definir el modelo lineal (ecuación de la recta)
f_x = m * x + b

# 4. Calcular la Suma de Cuadrados de los Errores (SSE)
sse = sum([(y_coords[i] - f_x.subs(x, x_coords[i]))**2 for i in range(len(x_coords))])

# 5. Calcular las derivadas parciales de SSE con respecto a m y b
diff_m = sympy.diff(sse, m)
diff_b = sympy.diff(sse, b)

# 6. Resolver el sistema de ecuaciones (derivadas parciales igualadas a cero)
solution_coeffs = sympy.solve([diff_m, diff_b], (m, b))

# 7. Extraer los valores de m y b
m_val = solution_coeffs[m]
b_val = solution_coeffs[b]

# 8. Construir la ecuación final de mínimos cuadrados
least_squares_equation = f_x.subs({m: m_val, b: b_val})

print(f"Datos X: {x_coords}")
print(f"Datos Y: {y_coords}")
print(f"Ecuación del modelo (f(x)): {f_x}")
print(f"Expresión de la Suma de Cuadrados de los Errores (SSE): {sse}")
print(f"Derivada parcial de SSE con respecto a m: {diff_m}")
print(f"Derivada parcial de SSE con respecto a b: {diff_b}")
print(f"Coeficientes encontrados (m, b): {solution_coeffs}")
print(f"La ecuación de mínimos cuadrados es: y = {least_squares_equation}")
```
